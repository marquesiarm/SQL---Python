```python
# importanto a biblioteca sqlite3
import sqlite3
```

```python
#Criando um banco de dados
Conexao = sqlite3.connect('Meu_banco_dados')
```

```python
#Craindo uma variável para chamar o banco de dados
Cursor = Conexao.cursor()
```

```python
#Criando uma tabela
Cursor.execute(
    'CREATE TABLE Minha_Tabela (Data text, nome text, idade real)'
)

Conexao.commit()
```

```python
#Inner join
Consulta = Cursor.execute(
    '''SELECT * FROM Tab_Vendas AS vendas
       INNER JOIN Tab_Vendedores AS vendor
       ON vendas.id = vendor.id
       
    '''
).fetchall()

for Linha in Consulta:
  print(Linha)
```

```python
Consulta = Cursor.execute(
    '''SELECT Nome, Vendas FROM Tab_Vendedores AS vendor
       INNER JOIN Tab_Vendas AS vendas
       ON vendor.id = vendas.id
    '''
).fetchall()

for Linha in Consulta:
  print(Linha)
```
